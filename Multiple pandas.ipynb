```python
countries_df = pd.read_csv('dictionary.csv')
countries_df.head()
```

```python
summer_df = pd.read_csv("summer.csv")
summer_df.head()
```

```python
summer_df = summer_df.rename(columns = {"Country": "Code"})
summer_df.head()
```

```python
winter_df = pd.read_csv('winter.csv')
winter_df.head()
```

```python
winter_df.rename(columns={"Country": "Code"}, inplace=True)
winter_df.head()
```

```python
summer_countries_df = summer_df.merge(countries_df, on='Code', how='inner')
summer_countries_df.head()
```

```python
summer_countries_df = summer_df.merge(countries_df, on="Code")
summer_countries_df["Season"] = "Summer"
summer_countries_df.head()
```

```python
winter_countries_df = winter_df.merge(countries_df, on="Code")
winter_countries_df["Season"] = "Winter"
winter_countries_df.head()
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], sort=False)
all_df
```

```python
len(all_df)
```

```python
all_df.sort_values("Year",ascending=False)
```

```python
all_recent_df = all_df[all_df["Year"]>=1984]
all_recent_df
```

```python
all_top_10_df = all_recent_df.groupby(["Country"])
all_top_10_df
```

```python
all_top_10_df= all_top_10_df[["Medal"]].count()
```
